```python
import pandas as pd
import numpy as np
import json
import joblib
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score
import os

# =====================================================
# üìå PATHS
# =====================================================
data_path = r"C:\Users\NXTWAVE\Downloads\Particale 2.5 Detection\archive\pollution_us_2000_2016.csv"
save_path = r"C:\Users\NXTWAVE\Downloads\Particale 2.5 Detection"

os.makedirs(save_path, exist_ok=True)

# =====================================================
# üìå STREAMING ONLINE MODEL
# =====================================================
model = SGDRegressor(max_iter=5, learning_rate="constant", eta0=0.001)

scaler = MinMaxScaler()
label_encoders = {"State": LabelEncoder(), "County": LabelEncoder(), "City": LabelEncoder()}

target = "NO2 Mean"
first_batch = True

last_chunk_X = None
last_chunk_y = None

# =====================================================
# üìå TRAIN MODEL CHUNK-BY-CHUNK (NO MEMORY ERROR)
# =====================================================
for chunk in pd.read_csv(data_path, chunksize=20000):

    # Drop high-cardinality text columns
    chunk = chunk.drop(columns=["Address", "NO2 Units", "O3 Units", "SO2 Units", "CO Units"])

    # Remove rows missing target
    chunk = chunk.dropna(subset=[target])

    # Date handling
    chunk["Date Local"] = pd.to_datetime(chunk["Date Local"])
    chunk["Year"] = chunk["Date Local"].dt.year
    chunk["Month"] = chunk["Date Local"].dt.month
    chunk["Day"] = chunk["Date Local"].dt.day
    chunk = chunk.drop(columns=["Date Local"])

    # Label encode
    for col in ["State", "County", "City"]:
        chunk[col] = label_encoders[col].fit_transform(chunk[col].astype(str))

    # IMPUTE missing values
    for col in chunk.columns:
        if chunk[col].dtype != "object":
            chunk[col] = chunk[col].fillna(chunk[col].mean())
        else:
            chunk[col] = chunk[col].fillna(chunk[col].mode()[0])

    # Split X and y
    y = chunk[target]
    X = chunk.drop(columns=[target])

    # Scale small chunks
    if first_batch:
        X_scaled = scaler.fit_transform(X)
        first_batch = False
    else:
        X_scaled = scaler.transform(X)

    # Online model update
    model.partial_fit(X_scaled, y)

    # Save last chunk for graphs
    last_chunk_X = X_scaled
    last_chunk_y = y

    print("Processed chunk:", len(chunk))


# =====================================================
# üìå SAVE MODEL + SCALER + ENCODERS
# =====================================================
joblib.dump(model, save_path + r"\pollution_model.pkl")
joblib.dump(scaler, save_path + r"\scaler.pkl")
joblib.dump(label_encoders, save_path + r"\label_encoders.pkl")

print("\nüéâ MODEL SAVED SUCCESSFULLY!")

# =====================================================
# üìå EVALUATE ON LAST CHUNK
# =====================================================
pred = model.predict(last_chunk_X)

mse = mean_squared_error(last_chunk_y, pred)
r2 = r2_score(last_chunk_y, pred)

print("\nMSE:", mse)
print("R2 Score:", r2)

# =====================================================
# üìå SAVE RESULT CSV & JSON
# =====================================================
result_df = pd.DataFrame({
    "Actual": last_chunk_y.values,
    "Predicted": pred
})

result_df.to_csv(save_path + r"\pollution_results.csv", index=False)

with open(save_path + r"\pollution_predictions.json", "w") as f:
    json.dump({
        "actual_values": last_chunk_y.values.tolist(),
        "predicted_values": pred.tolist(),
        "mse": float(mse),
        "r2": float(r2)
    }, f, indent=4)

print("\nüìÅ Saved: pollution_results.csv & pollution_predictions.json")

# =====================================================
# üìå GRAPH GENERATION + SAVE AS PNG
# =====================================================
errors = abs(last_chunk_y - pred)

# ---------------- GRAPH 1: Error Trend ----------------
plt.figure(figsize=(10,5))
plt.plot(errors[:1000])
plt.title("Error Trend")
plt.xlabel("Index")
plt.ylabel("Error")
plt.grid()
plt.savefig(save_path + r"\error_trend.png")
plt.show()

# ---------------- GRAPH 2: Scatter Plot ----------------
plt.figure(figsize=(6,6))
plt.scatter(last_chunk_y[:1000], pred[:1000], alpha=0.3)
plt.title("Actual vs Predicted Scatter Plot")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid()
plt.savefig(save_path + r"\scatter_plot.png")
plt.show()

# ---------------- GRAPH 3: Comparison Graph ----------------
plt.figure(figsize=(12,5))
plt.plot(last_chunk_y.values[:300], label="Actual", color="blue")
plt.plot(pred[:300], label="Predicted", color="red")
plt.title("Comparison Graph ‚Äî Actual vs Predicted")
plt.xlabel("Index")
plt.ylabel("NO2 Mean")
plt.legend()
plt.grid()
plt.savefig(save_path + r"\comparison_graph.png")
plt.show()

# ---------------- GRAPH 4: Residual Distribution ----------------
plt.figure(figsize=(10,5))
plt.hist(errors, bins=40, color='orange')
plt.title("Residual Error Distribution")
plt.xlabel("Error")
plt.ylabel("Frequency")
plt.grid()
plt.savefig(save_path + r"\residual_distribution.png")
plt.show()

print("\nüéâ ALL GRAPHS SAVED SUCCESSFULLY IN FOLDER:\n", save_path)
```
